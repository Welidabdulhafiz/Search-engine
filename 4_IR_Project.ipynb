```python
@anvil.server.callable
def stem(x):
    Word=convert_to_phonetic(x)
    if len(Word)>3:
        for x in Prefix:
            if Word.startswith(x):
                Word=Word.replace(x,"")
        for y in Sufix:
            if Word.endswith(y):
                Word=Word.replace(y,"")
    #print("This is the stemmed phonetic form")
            
    #print(Word)
    n =len(Word)
    i = 0
    last = ""
    while i<n:
        if i == n-1:
            check=Word[i]
        else:
            check=Word[i]+Word[i+1]
        if (vowel_check(check)) == 1:
            rev=Rev(check)
            last=last+rev
            i += 2
        else:
            last=last+Word[i]
            i += 1
    #print(last)
    return last
        
    
    
    
```
